```python
%matplotlib inline
import xarray as xr
import glob
import numpy as np
```

```python
fs = glob.glob('NDWI_composite/northern_NDWI_tile-*.tif')
fs.sort()

check_fs = []
for f in fs:
    data = xr.open_rasterio(f)
    bad_frac = np.isnan(data).mean().values
    if bad_frac>0: 
        print(f, bad_frac)
        check_fs.append(f)
```
